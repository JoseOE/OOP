```python
print(dict)
  
for valor in dict :
    print (valor)
  
for valor in dict.values() :
    print (valor)
  
print(dict['Nombre'])

print(dict.items())

for key, valor in dict.items() :
    print (key, valor)

dict['Nombre'] = 'Freyra'

for key, valor in dict.items() :
    print (key, valor)
    
print(dir(dict))
print(help(dict))
```
